```python
import cv2
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1=cv2.imread('/home/karandeep/Downloads/ueCwZ.png')
```

```python
output1=img1.copy()
gray=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
```

```python
circles=cv2.HoughCircles(gray,cv2.HOUGH_GRADIENT,1.2,67)

if circles is not None:
    circles=np.round(circles[0,:]).astype("int")
    
    for(x,y,r)in circles:
        
        cv2.circle(output1,(x,y),r,(0,255,0),4)
        cv2.rectangle(output1,(x-5,y-5), (x+5,y+5),(0,128,255),-1)
        
    cv2.imshow("output",np.hstack([img1,  output1]))
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
plt.imshow(img1)
```

```python
plt.imshow(output1)
```
